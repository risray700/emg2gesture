```python
import time
import pyttsx3


def finger_exercise_speech(repetitions=10, interval=5, speech_rate=180):
    """Prompts finger movements with speech at specified intervals."""

    exercises = [
        "Close your fingers in a fist, for 3 seconds",
        "Extend out all your fingers, for 8.85 seconds",
        "Close your fingers in a fist, for 3 seconds",
        "Extend out your thumb finger, keeping others closed, for 8.85 seconds",
        "Close your fingers in a fist, for 3 seconds",
        "Extend out your index finger, keeping others closed, for 8.85 seconds",
        "Close your fingers in a fist, for 3 seconds",
        "Extend out your middle finger, keeping others closed, for 8.85 seconds",
        "Close your fingers in a fist, for 3 seconds",
        "Extend out your ring and pinky finger, keeping others closed, for 8.85 seconds",
        
    
    ]

    engine = pyttsx3.init()  # Initialize the TTS engine
    engine.setProperty("rate", speech_rate)  # set new speech rate.

    for _ in range(repetitions):
        for exercise in exercises:
            print(exercise)  # Keep the print, as it is helpful to have a visual output.
            engine.say(exercise)  # Speak the exercise instruction
            engine.runAndWait()  # Wait until speech is finisheds
            time.sleep(interval)

    engine.say("Finger exercise complete!")
    engine.runAndWait()
    print("Finger exercise complete!")


if __name__ == "__main__":
    finger_exercise_speech()
```
